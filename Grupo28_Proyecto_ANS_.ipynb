## Revisión preliminar de antecedentes en la literatura

### Clasificación en la Industria Petroquímica

**1. Segmentación de Clientes en la Industria Petroquímica:**

La industria petroquímica atiende a una amplia gama de clientes, desde pequeñas empresas manufactureras hasta grandes corporaciones multinacionales. Estos clientes pueden diferir significativamente en términos de volumen de compras, frecuencia de pedidos, necesidades específicas de productos, y sensibilidad al precio. Además, los productos petroquímicos incluyen una variedad de materiales como plásticos, resinas, productos químicos especializados, y combustibles, cada uno con diferentes aplicaciones y mercados.

La diversidad de clientes y productos presenta un desafío significativo en la segmentación. Los métodos tradicionales de segmentación, que a menudo se basan en criterios demográficos, o de tamaño de empresa, suelen ser insuficientes para capturar las complejidades de este sector. Es por esto que las técnicas de clustering pueden aportar un valor significativo, permitiendo una segmentación más granular y precisa basada en patrones reales de comportamiento del cliente.

**2. Clustering para la Optimización de Precios:**

En la industria petroquímica, los precios pueden variar ampliamente según factores como el volumen de compra, la relación a largo plazo con el cliente, y las condiciones del mercado global. El clustering puede ayudar a identificar grupos de clientes con sensibilidad al precio similar o con patrones de compra similares, permitiendo a las empresas optimizar su estrategia de precios para maximizar los ingresos y la rentabilidad.

Estudios han demostrado que el clustering basado en el comportamiento de compra (como la frecuencia de pedidos, tamaño de pedidos y elasticidad del precio) puede llevar a una mejor comprensión de la disposición de los clientes a pagar, lo que a su vez permite la implementación de precios dinámicos o personalizados. 

**3. Clustering para la Mejora de la Cadena de Suministro:**

La industria petroquímica depende en gran medida de una cadena de suministro compleja, que incluye la producción, almacenamiento, y distribución de productos a nivel global. El clustering puede ser utilizado para optimizar la logística, identificando grupos de clientes con necesidades similares en términos de tiempos de entrega, métodos de transporte preferidos, y ubicaciones geográficas.

Agrupar clientes por proximidad geográfica o por similitudes en sus demandas puede permitir una planificación más eficiente de la cadena de suministro, reduciendo costos asociados con el transporte y almacenamiento. 

**4. Clustering para la Innovación en Productos:**

El análisis de clustering puede ser fundamental en la identificación de oportunidades para el desarrollo de nuevos productos o la personalización de productos existentes. Al segmentar a los clientes en función de sus necesidades específicas y patrones de consumo, las empresas petroquímicas pueden identificar nichos de mercado que no están siendo bien atendidos.

**5. Clustering en la Gestión del Riesgo:**

El clustering también puede ser utilizado para evaluar el riesgo asociado con diferentes grupos de clientes, especialmente en un contexto económico volátil. Identificar clientes con un alto riesgo de incumplimiento o que tienden a reducir sus pedidos en respuesta a fluctuaciones del mercado puede ayudar a las empresas a gestionar mejor su exposición al riesgo.


**6. Ejemplos Prácticos de Clasificacion en la Industria Petroquímica:

En la literatura, hay múltiples casos de estudio que demuestran la efectividad del clustering en la industria petroquímica. Por ejemplo, se ha utilizado clustering para identificar segmentos de clientes que valoran más la calidad del producto que el precio, lo que permite implementar estrategias de diferenciación de productos premium que permiten aumentar la lealtad del cliente y los márgenes de beneficio.

En cuanto a tecnologías y herramientas utilizadas: En la práctica, las empresas petroquímicas han utilizado herramientas como K-Means, análisis de componentes principales (PCA) combinado con clustering, y técnicas avanzadas de machine learning como DBSCAN para obtener insights accionables de grandes volúmenes de datos.

La aplicación de técnicas de clustering en la industria petroquímica tiene un potencial significativo para mejorar la segmentación de clientes, optimizar la cadena de suministro, desarrollar productos innovadores y gestionar riesgos de manera más efectiva. Al utilizar clustering, las empresas pueden descubrir patrones ocultos en los datos, lo que les permite tomar decisiones más informadas y estratégicas que pueden conducir a una ventaja competitiva en el mercado.

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
from sklearn.neighbors import NearestNeighbors
import numpy as np

# Seleccionar las variables relevantes
variables = [
    'Ingreso_USD',
    'Contribucion_USD'
]
df_subset = df[variables]

# Estandarizar las variables
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_subset)

# Encontrar el mejor valor para eps usando el gráfico de distancia de k-ésimo vecino más cercano
neigh = NearestNeighbors(n_neighbors=5)
neigh.fit(df_scaled)
distances, indices = neigh.kneighbors(df_scaled)

# Ordenar las distancias para crear el gráfico
distances = np.sort(distances[:, 4], axis=0)
plt.plot(distances)
plt.xlabel('Puntos ordenados')
plt.ylabel('Distancia')
plt.grid(True)
plt.title('Gráfico de distancia de k-ésimo vecino más cercano')
plt.show()
```


**Posible valor para eps:**
El codo del gráfico parece estar entre 0.5 y 2.0, pero debido a que el cambio es abrupto cerca del final a derecha, se recomendaría probar valores de eps en ese rango, alrededor de 1.0


```python

# Aplicar DBSCAN con los valores elegidos de eps y min_samples
eps = 1.0  # Ajusta este valor basado en el gráfico anterior
min_samples = 5
dbscan = DBSCAN(eps=eps, min_samples=min_samples)
clusters = dbscan.fit_predict(df_scaled)

# Añadir los resultados a los datos originales
df['Cluster'] = clusters

# Resumen de los clusters encontrados
cluster_summary = df.groupby('Cluster').agg({
    'Ingreso_USD': ['mean', 'std'],
    'Contribucion_USD': ['mean', 'std'],
    'SubFamilia': lambda x: x.mode()[0],
    'Empaque': lambda x: x.mode()[0],
    'LineaNegocio': lambda x: x.mode()[0],
    'Canal': lambda x: x.mode()[0],
    'PaisCorto': lambda x: x.mode()[0],
    'Region': lambda x: x.mode()[0],
    'Mercado': lambda x: x.mode()[0],
    'Segmento': lambda x: x.mode()[0]
}).reset_index()

print(cluster_summary)
print(f'Número de clusters encontrados: {len(cluster_summary)}')

# Guardar los resultados con las etiquetas de clusters
df.to_csv('clientes_segmentados_dbscan.csv', index=False)

print("Clustering con DBSCAN completado y resultados guardados en 'clientes_segmentados_dbscan.csv'.")


plt.figure(figsize=(14, 8))
sns.scatterplot(x=df['Ingreso_USD'], y=df['Contribucion_USD'], hue=df['Cluster'], palette='viridis', s=100, alpha=0.7)
plt.title('Segmentación de Clientes por Ingreso y Contribución (DBSCAN)')
plt.xlabel('Ingreso USD')
plt.ylabel('Contribución USD')
plt.grid(True)
plt.show()
```
